```python
%matplotlib inline
#Main directory contains data 
#mainDir = 'C:/Users/Tee/Documents/Active/EducationProject/SEOL/VideoAnalytics/main' #(Tee's ICME local machine)
mainDir = 'C:/Users/Admin/Dropbox/Active/EducationProject/SEOL/VideoAnalytics/main' #(Tee's dropbox machine)
#Here are directories that we will work with
OutputBigDirectory = mainDir + '/output/big/'+ coursename +'/'+ videoname 
OutputSmallDirectory = mainDir + '/output/small/'+ coursename +'/'+ videoname +'/SessionVisualize'
if not os.path.exists(OutputSmallDirectory):
    os.makedirs(OutputSmallDirectory)
```

Here we read out clean data and format timestamp. 

```python
#Do we want HashedUser, UserType and VideoActionTypeName to be string for sorting purpose now? Probably not.
CleanData = pd.read_csv(OutputBigDirectory+"/CleanData.csv",
                  dtype={'HashedUser':'object','UserType':'object','VideoActionTypeName':'object'})
#CleanSessionSummary = pd.read_csv(OutputBigDirectory+"/CleanSessionSummary.csv")
CleaningSummary = pd.read_csv(mainDir + '/output/small/'+ coursename +'/'+ videoname +"/CleaningSummary.csv")
MaxSecondMarker = int(CleaningSummary.iloc[4,1])
MaxMinuteMarker = int(CleaningSummary.iloc[5,1])

#convert ActionTime (aka. user timestamp) into datetime format
#add floating point if needed
CleanData.ActionTime[CleanData.ActionTime.map(lambda x: len(x)==19)] = \
    CleanData.ActionTime[CleanData.ActionTime.map(lambda x: len(x)==19)].map(lambda x: x+ '.000000')
CleanData.ActionTime =CleanData.ActionTime.map(lambda x: datetime.datetime.strptime(x,"%Y-%m-%d %H:%M:%S.%f"))
```

```python
#pull up some stats about sessions
[len(set(CleanData.HashedUser)),len(set(CleanData.SessionNumber))]
```

## Watching Session Plot Generation and Investigation

If (username, session number by user) or (session number) is known, one can query and make visualization as shown below. If not, we search through session numbers to find one that is presenting some distinct behavior. Below shows the generation of first 100 watching session plots. Note that there are various components which we will explain later.

```python
l = list(set(CleanData.SessionNumber))
for j in range(0,100):
    d = CleanData[CleanData.SessionNumber == l[j]]
    
    fig = plt.figure(figsize=(10,6))
    ax  = fig.add_subplot(111)
    # you can set the position manually, with setting left,buttom, width and height of the axis object
    ax.set_position([0.1,0.1,0.7,0.8]) #this is important as we added legend on the right

    #Loaded
    for i in map(lambda x: x.to_datetime(),list(d[d.VideoActionTypeName == 'Loaded'].ActionTime)):
        ax.axvline(i, color='green',linestyle='solid',label = 'Play')
    #End
    for i in map(lambda x: x.to_datetime(),list(d[d.VideoActionTypeName == 'End'].ActionTime)):
        ax.axvline(i, color='red',linestyle='solid',label= 'Pause')

    #Pause
    for i in map(lambda x: x.to_datetime(),list(d[d.VideoActionTypeName == 'Pause'].ActionTime)):
        ax.axvline(i, color='red',linestyle='solid', label = 'Pause')
    #Play
    for i in map(lambda x: x.to_datetime(),list(d[d.VideoActionTypeName == 'Play'].ActionTime)):
        ax.axvline(i, color='green',linestyle='solid', label = 'Play')

    #Minute Marker
    plt.plot(map(lambda x: x.to_datetime(),list(d[d.VideoActionTypeName == 'MinuteMarker'].ActionTime)),\
             d[d.VideoActionTypeName == 'MinuteMarker'].StartTime,'.',color='black',label='MinuteMarker')
    #Watching line
    for i in [i for i, elem in enumerate(d.StepType=='Watch') if elem]:
        ax.plot([d.ActionTime.iloc[i].to_datetime(), \
                  d.ActionTime.iloc[i].to_datetime()+ datetime.timedelta(seconds=d.UserStepLength.iloc[i])],\
                 [d.EndTime.iloc[i], d.EndTime.iloc[i]+d.VideoStepLength.iloc[i]],\
                 color='black', label= 'Watch')
    #Jump
    djump = d[d.VideoActionTypeName == 'Jump']
    for i in range(0,djump.shape[0]):
        ax.vlines(djump.ActionTime.iloc[i].to_datetime(),djump.StartTime.iloc[i] ,djump.EndTime.iloc[i],\
                   color='orange',linestyle='solid',label = 'Jump')

    #SpeedChange
    for i in map(lambda x: x.to_datetime(),list(d[d.VideoActionTypeName == 'SpeedChange'].ActionTime)):
        ax.axvline(i, color='blue',label='SpeedChange')

    ax.set_xlabel('Student Timeline')
    ax.set_xlim(d.ActionTime.iloc[0]-datetime.timedelta(seconds=5),d.ActionTime.iloc[-1]+datetime.timedelta(seconds=5))
    ax.set_ylabel('Video Location (s)')

    from collections import OrderedDict
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = OrderedDict(zip(labels, handles))
    ax.legend(by_label.values(), by_label.keys(), loc='center left', bbox_to_anchor=(1, 0.5),title='VideoActionTypeName')
    
    #save figure
    filename = 'SessionNumber_'+ str(j) +'.png'
    fig.savefig(os.path.join(OutputSmallDirectory, filename),dpi=100)
```

## Showcase of watching sessions visualization

Here are selected watching sessions. Note that we have remove components that might be confusing or not clearly visible. Each plot has horizontal axis representing student timeline, and vertical axis representing video timeline or location. Black color components indicate signal of watching, while other color components indicate signal of student's interaction with videos.

$\textbf{The first figure}$ shows watching session with multiple cases of pausing (red) and playing (green).

```python
l = list(set(CleanData.SessionNumber))
j=24
d = CleanData[CleanData.SessionNumber == l[j]]

fig = plt.figure(figsize=(10,6))
ax  = fig.add_subplot(111)
# you can set the position manually, with setting left,buttom, width and height of the axis object
ax.set_position([0.1,0.1,0.7,0.8]) #this is important as we added legend on the right

#Loaded
for i in map(lambda x: x.to_datetime(),list(d[d.VideoActionTypeName == 'Loaded'].ActionTime)):
    ax.axvline(i, color='green',linestyle='solid',label = 'Play')
#End
for i in map(lambda x: x.to_datetime(),list(d[d.VideoActionTypeName == 'End'].ActionTime)):
    ax.axvline(i, color='red',linestyle='solid',label= 'Pause')
    
#Pause
for i in map(lambda x: x.to_datetime(),list(d[d.VideoActionTypeName == 'Pause'].ActionTime)):
    ax.axvline(i, color='red',linestyle='solid', label = 'Pause')
#Play
for i in map(lambda x: x.to_datetime(),list(d[d.VideoActionTypeName == 'Play'].ActionTime)):
    ax.axvline(i, color='green',linestyle='solid', label = 'Play')

#Minute Marker
plt.plot(map(lambda x: x.to_datetime(),list(d[d.VideoActionTypeName == 'MinuteMarker'].ActionTime)),\
         d[d.VideoActionTypeName == 'MinuteMarker'].StartTime,'.',color='black',label='MinuteMarker')
#Watching line
for i in [i for i, elem in enumerate(d.StepType=='Watch') if elem]:
    ax.plot([d.ActionTime.iloc[i].to_datetime(), \
              d.ActionTime.iloc[i].to_datetime()+ datetime.timedelta(seconds=d.UserStepLength.iloc[i])],\
             [d.EndTime.iloc[i], d.EndTime.iloc[i]+d.VideoStepLength.iloc[i]],\
             color='black', label= 'Watch')
#Jump
djump = d[d.VideoActionTypeName == 'Jump']
for i in range(0,djump.shape[0]):
    ax.vlines(djump.ActionTime.iloc[i].to_datetime(),djump.StartTime.iloc[i] ,djump.EndTime.iloc[i],\
               color='orange',linestyle='solid',label = 'Jump')

#SpeedChange
for i in map(lambda x: x.to_datetime(),list(d[d.VideoActionTypeName == 'SpeedChange'].ActionTime)):
    ax.axvline(i, color='blue',label='SpeedChange')

ax.set_xlabel('Student Timeline')
ax.set_xlim(d.ActionTime.iloc[0]-datetime.timedelta(seconds=5),d.ActionTime.iloc[-1]+datetime.timedelta(seconds=5))
ax.set_ylabel('Video Location (s)')

from collections import OrderedDict
handles, labels = plt.gca().get_legend_handles_labels()
by_label = OrderedDict(zip(labels, handles))
ax.legend(by_label.values(), by_label.keys(), loc='center left', bbox_to_anchor=(1, 0.5),title='VideoActionTypeName')

#save figure
filename = 'SessionVisualize_A.png'
fig.savefig(os.path.join(OutputSmallDirectory, filename),dpi=fig.dpi)
fig.show()
```
